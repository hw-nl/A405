```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

pressures = np.array([1000,800,600]) #hPa
temp = np.arange(250,305,5) #K

def ln_theta(temp, pressure):
    """
    Input:  temperature, K
            pressure, hPa
    Output: ln(theta)
    """

    press0 = 100000 # Pa
    press = pressure*100
    Rd = 287  # J kg-1 K-1
    Cp = 1004  # J kg-1 K-1
    
    return np.log(temp) + Rd/Cp*(np.log(press0) - np.log(press))

lntheta_1 = ln_theta(temp, pressures[0])
lntheta_2 = ln_theta(temp, pressures[1])
lntheta_3 = ln_theta(temp, pressures[2])

fig,ax = plt.subplots(1,1)
ax.plot(temp,lntheta_1, label="1000 hPa")
ax.plot(temp,lntheta_2, label="800 hPa")
ax.plot(temp,lntheta_3, label="600 hPa")
ax.set(xlabel="T (K)",ylabel="ln theta",
      title = "Tephigram with isobar")
ax.grid(True)
ax.legend()
```
